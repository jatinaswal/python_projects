```python
# Diwali Sales Analysis – Business Insights Project

# Customer Purchase Behavior | Revenue Optimization | Festive Season Sales

# This project analyzes Diwali sales data to understand:
# - Who buys more? (Gender, Age, Marital Status, Occupation)
# - Where do most sales come from? (States, Zones)
# - What do customers prefer? (Product Categories, Orders)

# Tools & Libraries :-
# - Python (Pandas)
# Data Visualization: Matplotlib

# Goal: Help businesses optimize "marketing strategy, product inventory, and revenue" during festive seasons.


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("diwalisalesdata.csv", encoding="unicode_escape")
first5 = df.head()
print(first5)
print("----------------------------------------------------------------")

print("========= Data Info =========")
print(df.info())
print("----------------------------------------------------------------")

print("========= Missing Values =========")
print(df.isna().sum())
print("----------------------------------------------------------------")

print("========= Duplicate Values =========")
print(df.duplicated().sum())
print("----------------------------------------------------------------")

print("========= Statstical Summary =========")
print(df.describe())
print("----------------------------------------------------------------")

print("========= Data Types =========")
print(df.dtypes)
print("----------------------------------------------------------------")


df = df.drop(['Status', 'unnamed1'], axis=1)    #removing null columns.


df = df.drop_duplicates()   #droping duplicates values.


#converting following columns in numeric values.
df['Amount'] = pd.to_numeric(df['Amount'],errors = 'coerce')
df['Age'] = pd.to_numeric(df['Age'],errors = 'coerce')
df['Orders'] = pd.to_numeric(df['Orders'],errors = 'coerce')


df=df.dropna()   #dropping missing values.


df['Amount'] = df['Amount'].astype(int)   #changing datatype
df['Orders'] = df['Orders'].astype(int)
df['Age'] = df['Age'].astype(int)


#1. List all unique product categories.
unique_values = df['Product_Category'].unique()
print("========== Unique Product Category : ===========")
for i, values in enumerate(unique_values, start=1):
  print(f"{i}:{values}")
print("---------------------------------")


#2. Show total number of orders.
total_orders = df['Orders'].sum()
plt.figure(figsize=(3,3),facecolor='yellow')
plt.text(0.5, 0.4, f"Total Orders\n{total_orders}",
         ha='center', va='center', fontsize=20)
plt.box(False)
plt.xticks([])   # Remove ticks
plt.yticks([])
plt.show()
print("----------------------------------")


#3. Total number of users.
total_users = df['User_ID'].count()
plt.figure(figsize=(3,3),facecolor='skyblue')
plt.text(0.5,0.4,f"Total Users\n{total_users}",
         ha='center',va='bottom',fontsize=20,color='black')
plt.box(False)
plt.xticks([])   # Remove ticks
plt.yticks([])
plt.show()
print("-----------------------------------------------------------------------")


#4. Gender_counts
colors = ["lightgreen","skyblue"]
gender_counts = (df['Gender']
                .value_counts()
                .plot(kind = "bar",
                      color=colors,
                      figsize=(6,4),
                      title = "Gender Counts"
                     )
                )
for g in gender_counts.patches:
  gender_counts.text(
      g.get_x() + g.get_width()/2,
      g.get_height(),
      str(int(g.get_height())),
      va = "bottom",
      ha = "center"
  )
plt.xlabel("gender")
plt.ylabel("count")
plt.show()
print("\nINSIGHTS:")
print("• Female customers are more than male customers.")
print("• Women are more active in Diwali shopping.")
print("• Targeted marketing for females may increase revenue.")
print("-----------------------------------------------------------------------")


#5. Age_group_counts
age_group = (df['Age Group']
            .value_counts()
            .sort_index()
            .plot(
                kind='pie',
                autopct="%1.1f%%",
                fontsize = 8,
                startangle = 90
             )

)
plt.title("Age Group Distribution")
plt.ylabel("")
plt.show()
print("\nINSIGHTS:")
print("• Majority of the customers fall in the 26–35 age group.")
print("• 36–45 age group is the second major shopping population.")
print("• Customers above 45 contribute the least in Diwali shopping.")
print("• Target marketing should focus on the 26–45 age segment.")
print("-----------------------------------------------------------------------")


#6. State wise total customers.
colors = ["red", "blue", "green", "orange", "purple", "pink", "brown",
          "gray", "black", "yellow", "cyan", "magenta", "gold", "silver",
          "lime", "navy", "teal", "maroon"]
fig, ax = plt.subplots(figsize=(5,5), facecolor='papayawhip')
total_cus_state = df.groupby('State')['User_ID'].count().plot(kind = 'bar',
                                                               color=colors,
                                                               figsize=(7,5),
                                                               ax = ax)
for c in total_cus_state.patches:
  total_cus_state.text(
      c.get_x() + c.get_width()/2,
      c.get_height(),
      str(int(c.get_height())),
      va = 'bottom',
      ha = 'center'
  )
plt.title("No. of customers in each state",fontsize=15)
plt.xlabel('States')
plt.ylabel('Customers')
plt.show()
print("\nINSIGHTS:")
print("• Uttar Pradesh has the highest number of customers — biggest festive market opportunity.")
print("• Telangana has the lowest number of customers — needs targeted marketing campaigns.")
print("----------------------------------------------------------------------------")


#7. Find top 10 customers by purchase amount.
colors = ["red","blue","green","orange","purple","brown","pink","gray","cyan","olive"]

top_ten_cus = (df.groupby('Cust_name')['Amount']
              .sum()
              .sort_values(ascending=False)
              .head(10)
              .plot(kind = 'bar',
                    color=colors,
                    figsize = (8,5)
                    )
              )
for cus in top_ten_cus.patches:    #To show labels
  top_ten_cus.text(
      cus.get_x() + cus.get_width()/2,
      cus.get_height(),
      str(int(cus.get_height())),
      va = 'bottom',
      ha = 'center'
  )
plt.xlabel("Customer name")
plt.ylabel("Total Purchase Amount")
plt.title("Top 10 customers by purchase amount")
plt.show()
print("\nINSIGHTS:")
print("• Vishakha is the highest spender among all customers — key premium buyer.")
print("• Retaining top buyers like Vishakha can lead to strong long-term revenue growth.")
print("-----------------------------------------------------------------------------------------------")


#8. Male vs Female shoppings.
male_female_shopping = (df.groupby('Gender')['Amount']
                      .sum()
                      .plot(kind = 'pie',
                            autopct = "%1.1f%%",
                            colors=["#93C5FD", "#FBCFE8"],
                            figsize = (4,4),
                            startangle=90
                           )
                      )
plt.title("Male vs Female Shopping analysis", fontsize=12)
plt.ylabel("")  #removing y label
plt.show()
print("\nInsights:")
print("• Female customers contribute a higher share in total shopping amount.")
print("• Targeted offers toward female shoppers may result in higher sales.")
print("--------------------------------------------------")


#9. Which state contributed the highest revenue?
highest_revenue_state = df.groupby('State')['Amount'].sum().sort_values(ascending=False).head(1)

state_name = highest_revenue_state.index[0]    # Extracting clean values.
state_amount = highest_revenue_state.values[0]

plt.figure(figsize = (2.5,4),facecolor = 'cyan')
plt.text(
         0.5,0.4,
         f"Highest Revenue State :-\n \n{state_name}\n₹{state_amount}",
         ha = 'center',
         va='center',
         fontsize=20,
         color='black'
)
plt.box(False)
plt.xticks([])
plt.yticks([])
plt.show()
print("\nInsights:")
print(f"• The state contributing the highest revenue is: {state_name}")
print(f"• Total revenue generated from {state_name}: ₹{state_amount}")
print(f"• {state_name} is a key market for the business and should be prioritized for marketing and sales efforts.")
print("------------------------------------------------------------------------------------------")


#10. Which gender & age group combination buys more?
gender_age_revenue = df.groupby(['Gender', 'Age Group'])['Amount'].sum()

ax = gender_age_revenue.unstack().plot(kind="bar", figsize=(8,5))
for gender in ax.patches:
  ax.text(
      gender.get_x() + gender.get_width()/2,
      gender.get_height(),
      str(int(gender.get_height())),
      ha = 'center',
      va = 'bottom',
      fontsize = 6
  )
plt.title("Gender & Age Group Revenue Comparison")
plt.xlabel("Gender")
plt.ylabel("Total Amount")
plt.legend(title="Age Group")
plt.show()
print("\nINSIGHTS:")
print("• The Female (F) 26-35 age-group is the highest-spending segment, generating a total revenue of 30,945,638.")
print("• This group should be a primary focus for marketing and retention efforts.")
print("-----------------------------------------------------------------------------------------")


#11. Married vs Unmarried
colors = ["#A5B4FC", "#F9A8D4"]
marital_counts = (df.groupby('Gender')['Marital_Status']
                 .value_counts().unstack().plot(
                                      kind='bar',
                                      figsize=(5,4),
                                      color=colors
                                      )
)
for mc in marital_counts.patches:
  marital_counts.text(
      mc.get_x() + mc.get_width()/2,
      mc.get_height(),
      str(int(mc.get_height())),
      va = 'bottom',
      ha = 'center'
)
plt.title("Married vs Unmarried Customers")
plt.xlabel("Marital Status")
plt.ylabel("Count")
plt.show()
print("\nINSIGHTS:")
print("• Unmarried (Marital_Status=0) female customers form the largest customer group, with a count of 4,572.")
print("• For both genders the unmarried segment (Marital_Status=0) significantly outnumbers the married segment (Marital_Status=1).")
print("-------------------------------------------------------------------------------------------")


#12. total no. of order from top 10 states.
colors = [
    "#6A5ACD",  # Slate Purple
    "#20B2AA",  # Light Sea Green
    "#FFB347",  # Soft Orange
    "#C71585",  # Deep Pink Red
    "#8FBC8F",  # Sage Green
    "#708090",  # Slate Gray
    "#FFD700",  # Gold Yellow
    "#4682B4",  # Steel Blue
    "#FF7F50",  # Coral
    "#40E0D0"   # Turquoise
]
top_states_orders = (df.groupby('State')['Orders']
                     .sum()
                     .sort_values(ascending = False)
                     .head(10)
                     .plot(kind = 'bar',
                           figsize = (8,5),
                           title = "Top 10 states by no. orders",
                           color = colors
                          )
                    )
for ts in top_states_orders.patches:
  top_states_orders.text(
      ts.get_x() + ts.get_width()/2,
      ts.get_height(),
      str(int(ts.get_height())),
      va = 'bottom',
      ha = 'center'
)
plt.xlabel("States")
plt.ylabel("No. of orders")
plt.show()
print("\nINSIGHTS:")
print("• Uttar Pradesh is contributing the highest number of orders at 4,803.")
print("• The top 3 states (Uttar Pradesh, Maharashtra, and Karnataka) demonstrating a strong regional concentration of demand.")
print("• The remaining seven states contribute significantly less individually compared to the top three.")
print("-----------------------------------------------------------------------------")


#13.which agegroup did more shopping.
colors = [
    "#7D5BA6",  # Soft Violet
    "#F7A072",  # Peach Sand
    "#57C4E5",  # Sky Aqua
    "#9CE37D",  # Fresh Green
    "#F9D371",  # Warm Yellow
    "#E26D5A",  # Coral Brick
    "#A3A1F7"   # Mist Lavender
]
age_group_shopping = (df.groupby('Age Group')['Amount']
                      .sum()
                      .plot(
                          kind = 'bar',
                          title = 'Age Group-wise Shopping Analysis',
                          color = colors,
                          figsize = (5,5)
                         )
)
for ag in age_group_shopping.patches:
  age_group_shopping.text(
      ag.get_x() + ag.get_width()/2,
      ag.get_height(),
      str(int(ag.get_height())),
      va = 'bottom',
      ha = 'center',
      fontsize = 8
)
plt.xlabel("Age Group")
plt.ylabel("Total Amount")
plt.show()
print("\nINSIGHTS:")
print("• The 26-35 age group is the highest-spending segment with massive total revenue of 42,581,767.")
print("• The second highest-spending group is 36-45 with 22,135,135 in total purchases.")
print("• The 26-35 segment should be the primary market target.")
print("------------------------------------------------------------")


#14. top 5 selling product_category.
colors = [
    "#2EC4B6",  # Aqua Teal
    "#FF9F1C",  # Amber Orange
    "#CBF3F0",  # Mint Breeze
    "#E71D36",  # Crimson Red
    "#6C5B7B"   # Dusty Purple
]
top_selling_product = (df.groupby('Product_Category')['Amount']
                       .sum()
                       .sort_values(ascending = False)
                       .head(5)
                       .plot(
                           kind = 'bar',
                           title = "Top 5 Selling Product-Category",
                           figsize = (5,5),
                           color = colors
                       )
)
for tsp in top_selling_product.patches:
  top_selling_product.text(
      tsp.get_x() + tsp.get_width()/2,
      tsp.get_height(),
      str(int(tsp.get_height())),
      va = 'bottom',
      ha = 'center',
      fontsize = 8
)
plt.xlabel("Product Category")
plt.ylabel("Amount")
plt.show()
print("\nINSIGHTS:")
print("• Food is the highest-grossing category with 33,933,883 total revenue.")
print("• This suggests high frequency of purchase or high average order value for food items.")
print("-----------------------------------------------------------")


#15. dept with highest no. of buyers.
colors = [
    "#1A6F4C",  # Deep Emerald Green
    "#F6BD60",  # Soft Gold
    "#4C4C9D",  # Indigo Blue
    "#EE6055",  # Coral Red
    "#60D394",  # Fresh Green
    "#AAF683",  # Lime Mint
    "#FFD97D",  # Pastel Yellow
    "#FF9B85",  # Peach Orange
    "#5C80BC",  # Steel Blue
    "#C8553D",  # Rust Orange
    "#3A86FF",  # Bright Sky Blue
    "#8338EC",  # Vivid Purple
    "#FB5607",  # Electric Orange
    "#FFBE0B",  # Vibrant Yellow
    "#9B5DE5"   # Purple Orchid
]
dept_buyers = (df.groupby('Occupation')['User_ID']
               .nunique()
               .sort_values(ascending = False)
               .plot(
                   kind = 'bar',
                   title = "Departments With Highest No. Of Buyers",
                   figsize = (10,5),
                   color = colors
                  )
)
for db in dept_buyers.patches:
  dept_buyers.text(
      db.get_x() + db.get_width()/2,
      db.get_height(),
      str(int(db.get_height())),
      va = 'bottom',
      ha = 'center',
      fontsize = 8
)
plt.xlabel("Department")
plt.ylabel("No. Of Users")
plt.show()
print("\nINSIGHTS:")
print("• The IT Sector is the largest customer base with 1,125 buyers.")
print("• The Agriculture sector has the lowest number of buyers among the listed departments, with only **253 users**, suggesting it is a non-core market segment.")


# Final Note :-
# This analysis provides clear pathways to:
# - Grow festive season revenue.
# - Improve conversion through targeted marketing.
# - Optimize inventory and product mix.
```
